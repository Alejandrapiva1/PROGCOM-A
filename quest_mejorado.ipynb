```python
import ipywidgets as widgets
from IPython.display import display, clear_output


reglas = {
    "piedra": ["tijeras", "lagarto", "fuego"],
    "papel": ["piedra", "agua", "aire"],
    "tijeras": ["papel", "lagarto", "aire"],
    "spock": ["tijeras", "piedra", "rayo"],
    "lagarto": ["spock", "papel", "agua"],
    "fuego": ["papel", "lagarto", "aire"],
    "agua": ["fuego", "piedra", "rayo"],
    "aire": ["agua", "spock", "rayo"],
    "rayo": ["tijeras", "fuego", "piedra"]
}
opciones = list(reglas.keys())


puntos1 = 0
puntos2 = 0

nombre1 = widgets.Text(placeholder='Nombre Jugador 1')
nombre2 = widgets.Text(placeholder='Nombre Jugador 2')

eleccion1 = widgets.Password(placeholder="Escribe tu opciÃ³n (en secreto)", description="Jugador 1:")
eleccion2 = widgets.Password(placeholder="Escribe tu opciÃ³n (en secreto)", description="Jugador 2:")

jugar_btn = widgets.Button(description="Â¡Jugar ronda!", button_style="success")
reiniciar_btn = widgets.Button(description="Reiniciar partida", button_style="danger")

resultado_out = widgets.Output()
puntaje_out = widgets.Output()


def jugar_ronda(b):
    global puntos1, puntos2
    with resultado_out:
        clear_output()
        j1 = eleccion1.value.strip().lower()
        j2 = eleccion2.value.strip().lower()
        n1 = nombre1.value.strip() or "Jugador 1"
        n2 = nombre2.value.strip() or "Jugador 2"

        if j1 not in opciones or j2 not in opciones:
            print("â— Ambos jugadores deben escribir una opciÃ³n vÃ¡lida.")
            return

        print(f"{n1} eligiÃ³: {j1}")
        print(f"{n2} eligiÃ³: {j2}")

        if j1 == j2:
            print("ğŸŸ° Â¡Empate!")
        elif j2 in reglas[j1]:
            print(f"âœ… Â¡{n1} gana esta ronda!")
            puntos1 += 1
        else:
            print(f"âœ… Â¡{n2} gana esta ronda!")
            puntos2 += 1

        actualizar_puntaje()

def actualizar_puntaje():
    with puntaje_out:
        clear_output()
        n1 = nombre1.value.strip() or "Jugador 1"
        n2 = nombre2.value.strip() or "Jugador 2"
        print(f"ğŸ† Puntaje actual:")
        print(f"{n1}: {puntos1}")
        print(f"{n2}: {puntos2}")


def reiniciar_juego(b):
    global puntos1, puntos2
    puntos1 = 0
    puntos2 = 0
    eleccion1.value = ""
    eleccion2.value = ""
    resultado_out.clear_output()
    actualizar_puntaje()


jugar_btn.on_click(jugar_ronda)
reiniciar_btn.on_click(reiniciar_juego)


display(widgets.VBox([
    widgets.Label("ğŸ® Ingreso de nombres:"),
    nombre1,
    nombre2,
    widgets.Label("ğŸ”’ ElecciÃ³n privada (escribe una de estas: piedra, papel, tijeras, spock, lagarto, fuego, agua, aire, rayo):"),
    eleccion1,
    eleccion2,
    widgets.HBox([jugar_btn, reiniciar_btn]),
    puntaje_out,
    resultado_out
]))

```
