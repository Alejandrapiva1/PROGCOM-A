```python
from abc import ABC, abstractmethod
import ipywidgets as widgets
from IPython.display import display, clear_output


class PersonaCredito(ABC):
    def __init__(self, edad, tiene_tarjeta, sin_morosidad, contrato_fijo,
                 contrato_estable, tuvo_credito_otro_banco, tuvo_credito_aca):
        self.edad = edad
        self.tiene_tarjeta = tiene_tarjeta
        self.sin_morosidad = sin_morosidad
        self.contrato_fijo = contrato_fijo
        self.contrato_estable = contrato_estable
        self.tuvo_credito_otro_banco = tuvo_credito_otro_banco
        self.tuvo_credito_aca = tuvo_credito_aca

    @abstractmethod
    def calcular_puntos(self):
        pass

    @abstractmethod
    def aprobar_credito(self):
        pass


class SistemaEvaluacion(PersonaCredito):
    def calcular_puntos(self):
        puntos = 0
        if self.tiene_tarjeta:
            puntos += 2
        if self.sin_morosidad:
            puntos += 3
        if self.contrato_fijo:
            puntos += 2
        if self.contrato_estable:
            puntos += 3
        if self.tuvo_credito_otro_banco:
            puntos += 2
        if self.tuvo_credito_aca:
            puntos += 3

        if 18 <= self.edad <= 24:
            puntos += 2
        elif 25 <= self.edad <= 35:
            puntos += 4
        elif 36 <= self.edad <= 50:
            puntos += 3
        else:
            puntos += 1

        return puntos

    def aprobar_credito(self):
        return self.calcular_puntos() >= 7


edad_input = widgets.IntSlider(description='Edad', value=25, min=18, max=70)

tarjeta_chk = widgets.Checkbox(description='¿Tiene tarjeta otra franquicia?')
morosidad_chk = widgets.Checkbox(description='¿No está reportado negativamente?')
contrato_fijo_chk = widgets.Checkbox(description='¿Contrato fijo?')
contrato_estable_chk = widgets.Checkbox(description='¿Contrato indefinido?')
credito_otra_chk = widgets.Checkbox(description='¿Tuvo crédito otro banco?')
credito_aca_chk = widgets.Checkbox(description='¿Tuvo crédito aquí?')

boton = widgets.Button(description='Evaluar crédito', button_style='success')
salida = widgets.Output()


def evaluar_credito(b):
    with salida:
        clear_output()
        persona = SistemaEvaluacion(
            edad_input.value,
            tarjeta_chk.value,
            morosidad_chk.value,
            contrato_fijo_chk.value,
            contrato_estable_chk.value,
            credito_otra_chk.value,
            credito_aca_chk.value
        )
        puntos = persona.calcular_puntos()
        if persona.aprobar_credito():
            print(f"✅ Crédito aprobado (Puntaje: {puntos})")
        else:
            print(f"❌ Crédito no aprobado (Puntaje: {puntos})")


boton.on_click(evaluar_credito)


display(
    edad_input,
    tarjeta_chk,
    morosidad_chk,
    contrato_fijo_chk,
    contrato_estable_chk,
    credito_otra_chk,
    credito_aca_chk,
    boton,
    salida
)
```
