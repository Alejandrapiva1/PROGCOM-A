```python

flores = {
    'Rosa': {'precio': 10, 'cantidad': 30},
    'Tulipán': {'precio': 8, 'cantidad': 25},
    'Lirio': {'precio': 12, 'cantidad': 15},
    'Margarita': {'precio': 5, 'cantidad': 50},
    'Orquídea': {'precio': 20, 'cantidad': 10},
    'Clavel': {'precio': 7, 'cantidad': 40},
    'Gerbera': {'precio': 6, 'cantidad': 35},
    'Girasol': {'precio': 4, 'cantidad': 60},
    'Violeta': {'precio': 9, 'cantidad': 20},
    'Crisántemo': {'precio': 11, 'cantidad': 18},
    'Peonía': {'precio': 15, 'cantidad': 8},
    'Azucena': {'precio': 14, 'cantidad': 22},
    'Dalia': {'precio': 13, 'cantidad': 14},
    'Aster': {'precio': 16, 'cantidad': 12},
    'Cala': {'precio': 17, 'cantidad': 7},
    'Jazmín': {'precio': 18, 'cantidad': 5},
    'Gardenia': {'precio': 19, 'cantidad': 4},
    'Freesia': {'precio': 8, 'cantidad': 45},
    'Alstroemeria': {'precio': 6, 'cantidad': 55}
}


def comprar(flor, cantidad):
    if flor in flores:
        if flores[flor]['cantidad'] >= cantidad:
            flores[flor]['cantidad'] -= cantidad
            total = flores[flor]['precio'] * cantidad
            return total
        else:
            return "No hay flores suficientes."
    else:
        return "Flor no disponible."

def mostrar_inventario():
    print("\nInventario de flores:")
    for flor, detalles in flores.items():
        print(f"{flor}: {detalles['cantidad']} disponibles")

def mostrar_ganancias(ganancias):
    print(f"\nGanancias del día: ${ganancias}")


def clasificar_por_ventas(ventas):
    print("\nClasificación de flores vendidas (de mayor a menor):")

    lista_ventas = list(ventas.items())

    for i in range(len(lista_ventas)):
        for j in range(i + 1, len(lista_ventas)):
            if lista_ventas[i][1] < lista_ventas[j][1]:

                lista_ventas[i], lista_ventas[j] = lista_ventas[j], lista_ventas[i]

    for flor, cantidad in lista_ventas:
        print(f"{flor}: {cantidad} unidades")


def floristeria():
    ganancias = 0
    ventas = {}

    while True:
        mostrar_inventario()

        flor = input("\n¿Qué flor deseas comprar? (Escribe 'cerrar' para salir): ").capitalize()

        if flor == 'Cerrar':
            break

        cantidad = int(input(f"¿Cuántas {flor}s deseas comprar?: "))

        resultado = comprar(flor, cantidad)

        if isinstance(resultado, int):
            ganancias += resultado
            if flor not in ventas:
                ventas[flor] = 0
            ventas[flor] += cantidad
            print(f"¡Compra exitosa! Has comprado {cantidad} {flor}s por ${resultado}.")
        else:
            print(resultado)

    mostrar_ganancias(ganancias)
    clasificar_por_ventas(ventas)


floristeria()
```
